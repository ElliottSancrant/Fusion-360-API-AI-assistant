```python
import re

# Function to clean the "Description" field more aggressively
def clean_description(text):
    if pd.isna(text):
        return text  
    text = text.lower()
    text = re.sub(r'defined in namespace.*?;', '', text)
    text = re.sub(r'namespace \w+', '', text)
    text = re.sub(r'header file is .*?;', '', text)
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
df['Description'] = df['Description'].apply(clean_description)

df[['Name', 'Description']].head(25)
```
